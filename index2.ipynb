# Data Transformation

The square foot columns I log transformed and normalized. I one-hot encoded the categorical features of bedrooms, grade, and zipcode.

```python
# log transform sqft_lot, sqft_living
import numpy as np
log_sqft_lot = np.log(df_new['sqft_lot'])
log_sqft_living = np.log(df_new['sqft_living'])
log_price= np.log(df_new['price'])
```

```python
# normalize sqft_living and lot
norm_sqft_living = (log_sqft_living - log_sqft_living.mean())/log_sqft_living.std()

norm_sqft_lot = (log_sqft_lot - log_sqft_lot.mean())/log_sqft_lot.std()

norm_price = (log_price - log_price.mean())/log_price.std()

plt.hist(log_sqft_living, bins=20)
```

I one-hot encoded the categorical features.

```python
# get dummies
bedrooms_dummies = pd.get_dummies(df_new['bedrooms'], prefix='bdr', drop_first=True)
grade_dummies = pd.get_dummies(df_new['grade'], prefix='gr', drop_first=True)
zipcode_dummies = pd.get_dummies(df_new['zipcode'], prefix='zip', drop_first=True)

```

Finally, I put all of these features in a new dataframe.

```python
# concatenate final database
df_temp = pd.DataFrame([])
df_temp['norm_price'] = norm_price
df_temp['norm_sqft_living'] = norm_sqft_living
df_temp['norm_sqft_lot'] = norm_sqft_lot
df_final = pd.concat([df_temp, df_new['bedrooms'], df_new['grade'], zipcode_dummies], axis=1)
df_final.rename(columns = {0:'price'}, inplace = True)
df_final.head()
```

# Building the Model



One of the first considerations in buying a home is the square feet of the living area, so I did a quick linear model with just that feature.

```python
# create a linear model with statsmodel using just sqft_living

import statsmodels.api as sm
from statsmodels.formula.api import ols

formula = 'norm_price ~ norm_sqft_living'
model = ols(formula=formula, data=df_final).fit()
model.summary()
```

It's not a great model, but that's to be expected, as it's only one feature. Then I added in sqft_lot, bedrooms, grade, and zipcode to see how it improved the model.

```python
# create a linear model with statsmodel using sqft_living, sqft_lot, bedrooms, grade, and zipcode

outcome = 'norm_price'
predictors = df_final.drop('norm_price', axis=1)
pred_sum = '+'.join(predictors.columns)
formula = outcome + '~' + pred_sum

model = ols(formula=formula, data=df_final).fit()
model.summary()
```

This is a much improved model. The R-squared value is 0.768 and the Adjusted R-squared value is 0.767.

While this R-squared value may not be considered great for other scenarios, prices of individual homes are based a large number of factors, some which are emotional, and many of which are not included in this dataset (quality of schools, curb appeal, etc.). This model's primary value is in identifying the features that have the biggest influence on price. 

The primary value of this model would be a realtor using it to set an initial price for a home being placed on the market and then adjusting the price for the specific features of that home (any view, waterfront access, etc.).

However, several of the zipcodes have p-values of above 0.05. I then asked the question, what if we treated these zipcodes as outliers and removed them from the dataset? (Zipcodes not being ordinal, the scatterplot of zipcodes vs. price did not help identify outliers).

I then created a new model from the resulting dataframe.

```python
# log transform sqft_lot, sqft_living
import numpy as np
log_sqft_lot = np.log(df_revised['sqft_lot'])
log_sqft_living = np.log(df_revised['sqft_living'])
log_price = np.log(df_revised['price'])

# normalize sqft_living and lot
norm_sqft_living = (log_sqft_living - log_sqft_living.mean())/log_sqft_living.std()

norm_sqft_lot = (log_sqft_lot - log_sqft_lot.mean())/log_sqft_lot.std()

norm_price = (log_price - log_price.mean())/log_price.std()

# get dummies
zipcode_dummies = pd.get_dummies(df_revised['zipcode'], prefix='zip', drop_first=True)

# concatenate final database
df_temp = pd.DataFrame([])
df_temp['norm_price'] = norm_price
df_temp['norm_sqft_living'] = norm_sqft_living
df_temp['norm_sqft_lot'] = norm_sqft_lot
df_revised_final = pd.concat([df_temp, df_revised['bedrooms'], df_revised['grade'], zipcode_dummies], axis=1)
df_revised_final.rename(columns = {0:'price'}, inplace = True)

outcome = 'norm_price'
predictors = df_revised_final.drop('norm_price', axis=1)
pred_sum = '+'.join(predictors.columns)
formula = outcome + '~' + pred_sum

model = ols(formula=formula, data=df_revised_final).fit()
model.summary()
```

The result is a model with similar R-squared and Adjusted R-squared values, 0.761 and 0.760, respectively. It also brings the p-values down below 0.05, making the model significant.

The question then becomes, what reasoning is there for separating these homes from the rest of the dataset? For that, we need to understand the local geography.

![image info](./mapKingCounty.gif)

The zipcodes are in three cities: Enumclaw, Kent, Auburn, and Federal Way. 

Enumclaw - southeast corner of King County

It is the last city one drives through before reaching the north entrance of Mount Rainier National Park, and  a large part of its zipcode is very sparsely populated. It's reasonable that the park might be influencing prices compared to other small cities near rural areas (such as Duvall and Carnation, which are also closer to Seattle and make for a reasonable commute).

Kent and Auburn - south central King County  
Federal Way - southwest corner of King County

These cities have a different uniqueness, as they are on major commuting routes. Kent and Auburn are along State Route 167, and Federal Way is along I-5. These cities are as far south as most people in the county are willing to live and still commute north to Seattle or Bellevue/Redmond, the two biggest job centers in the county. Federal Way is a bit further south than Kent, and that make sense when you consider that 167 gets more conjested than I-5 during commuting hours. The commute times from Seattle and Bellevue/Redmond are about equivalent.

They are also close enough to Tacoma that one could commute to that major city. However, anything north of Federal Way and Kent would typically be too expensive for commuters on a Tacoma salary.

What this model seems to suggest is that Kent, Auburn, and Federal Way home prices have different influences not predicted by this model. A reasonable influence could be proximity to both the Seattle and Tacoma job markets. Since Tacoma is in Pierce County, modeling this influence is beyond the scope of this project that is limited to homes in King County.
